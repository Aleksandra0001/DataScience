# Ukraine birth

```python
url="https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8#%D0%9D%D0%B0%D1%80%D0%BE%D0%B4%D0%B6%D1%83%D0%B2%D0%B0%D0%BD%D1%96%D1%81%D1%82%D1%8C"
data_birth = pd.read_html(url, match="Коефіцієнт народжуваності в регіонах України", thousands=".", decimal="," )[0]
data_birth
```

```python
data_birth.plot(x='Регіон', y='1950', c='#1f77b4', alpha=0.5)
plt.xlabel('Region', fontsize='x-large', color='black')
plt.ylabel('Year', fontsize='x-large', color='black')
plt.title('Birth rate by the region', fontsize=15)
plt.xticks(rotation = 45)
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.show()
```

```python
second=data_birth.plot(
    x='Регіон', 
    y='2000', 
    linestyle="--", 
    color="#FF5733", 
    label="2000 year")
sns.set(rc={'figure.figsize':(11.7,8.27)})
plt.xticks(rotation = 45);
```

```python
data_birth.replace("—", np.NaN, inplace=True)
data_birth.iloc[:,-2:]=data_birth.iloc[:,-2:].astype(float)
data_change=data_birth.fillna(data_birth.mean(numeric_only=True).round(1))
last_row=data_change.shape[0]-1
data_change.drop(last_row,axis=0, inplace=True)
data_change[['1950','1960','1970','1990','2000','2012','2014','2019', 'Регіон']].max()
```

```python
labels=['1950','1960','1970','1990','2000','2012','2014','2019']
data=[31,27,21,17,12,16,15,11]
explode=[0.15, 0, 0, 0, 0, 0, 0, 0]
plt.pie(
    data,
    labels=labels,
    shadow=True,
    explode=explode,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35,
);
```

# Books

```python
url='https://drive.google.com/file/d/1MTlnE7r9gN-GQXXNnrGq45VnuYRhYhgd/view?usp=share_link'
url='https://drive.google.com/uc?id='+ url.split('/')[-2]
data_books = pd.read_csv(url)
data_books.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
data_books.head()
```

```python
sns.set(rc={'figure.figsize':(20,9)})
sns.swarmplot(x='genre', y='user_rating', data=data_books);
```

```python
sns.stripplot(x='year', y='reviews', hue='genre', data=data_books);
```

```python
top_50=data_books[:50]
# sns.set(rc={'figure.figsize':(15,9)})
table_4=sns.stripplot(x='author', y='price', hue='year', data=top_50)
table_4.set_xticks(range(len(top_50)))
table_4.set_xticklabels(table_4.get_xticklabels(), rotation=90);
```
